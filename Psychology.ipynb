```python
import pandas as pd
import streamlit as st

# Завантаження даних із згенерованого файлу
file_path = 'randomized_questions.csv'
data = pd.read_csv(file_path)

def run_interactive_test(data):
    questions = data["Питання"].unique()
    answers = {}

    st.title("Психологічний тест")

    if 'question_idx' not in st.session_state:
        st.session_state['question_idx'] = 0

    def calculate_results():
        scores = {}
        for question, selected_option in answers.items():
            option_weights = data[(data["Питання"] == question) & (data["Варіанти__відповідей"] == selected_option)].iloc[0, 2:]
            for category, weight in option_weights.items():
                scores[category] = scores.get(category, 0) + weight

        max_category = max(scores, key=scores.get)

        st.subheader("Результати тесту:")
        st.write(f"Найбільш вагома категорія: {max_category}")
        st.write("Скоринг по всіх категоріях:")
        for category, score in scores.items():
            st.write(f"{category}: {score}")

    if st.session_state['question_idx'] < len(questions):
        question = questions[st.session_state['question_idx']]
        options = data[data["Питання"] == question]["Варіанти__відповідей"].tolist()

        st.subheader(f"{question}")
        selected_option = st.radio("Оберіть варіант:", options)

        if st.button("Далі"):
            answers[question] = selected_option
            st.session_state['question_idx'] += 1
            st.experimental_rerun()
    else:
        calculate_results()

# Запуск інтерактивного тесту
run_interactive_test(data)
```
